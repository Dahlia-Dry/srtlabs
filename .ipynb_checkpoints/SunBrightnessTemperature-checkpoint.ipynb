# SRT Lab 1: The Sun's Brightness Temperature

```python
from dataprocessing import SRTData
import matplotlib.pyplot as plt
import digital_rf
import pandas as pd
import numpy as np
```

## Part 1: Npoint Scan

```python
#process raw SRT data to power time series
#alt = np.arange(30,51,step=5)
#az = np.arange(30,51,step=5)
srtdata = SRTData(datadir = '/Users/dahliadry/Projects/HaystackSummer2020/ch1/',
                        params=['power'], start=0, end=25).process()
```

```python
#make contour plot
fig, ax = plt.subplots()
x, y = np.meshgrid(np.asarray(srtdata['alt']),np.asarray(srtdata['az']))
z = np.asarray(srtdata['power']).reshape((5,5))
ax.set_xlabel('Altitude (degrees)')
ax.set_ylabel('Azimuth (degrees)')
ax.set_title('5x5 Npoint Scan')
brightness = ax.contour(x, y, z)
plt.show()
```

## Part 2: Azimuth/Altitude Scan

```python
#Altitude Scan
#Note to self -- look up how to convert power to temp
fig, ax = plt.subplots()
alt = np.arange(0,25)
power = np.asarray(srtdata['power'])
ax.set_xlabel('Altitude (degrees)')
ax.set_ylabel('Power')
ax.set_title('Altitude Scan')
ax.scatter(alt, power)
plt.show()
```
